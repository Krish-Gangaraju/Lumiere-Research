## CLEVR Ops - Answer Distribution

```python
ops_val = pd.read_csv('FinalQuestionsValidation.csv')
ops_val
```

```python
ops_val_values = ops_val["New Answer"].tolist()
count = 0
for i in ops_val_values:
    if (len(i) == 1):
        ops_val_values[count] = int(ops_val_values[count])
    elif (i[1] == "/"):
        ops_val_values[count] = 0
    elif (i[1] == "%"):
        ops_val_values[count] = 0
    else:
        ops_val_values[count] = float(ops_val_values[count])
    count += 1
len(ops_val_values)
```

```python
h = set(ops_val_values)
print(len(h))
ops_val_y = []
for x in h:
    ops_val_y.append(ops_val_values.count(x))
ops_val_y
print(len(ops_val_y))
```

```python
ops_val_x = h
len(ops_val_x)
```

```python
sum(ops_val_y)
```

```python
print(ops_val_x)
```

```python
print(ops_val_y)
```

```python
ops_histo_val = pd.DataFrame()
ops_histo_val["X"] = list(ops_val_x)
ops_histo_val["Y"] = ops_val_y
```

```python
ops_histo_val.to_excel("ops_val.xlsx")
```
