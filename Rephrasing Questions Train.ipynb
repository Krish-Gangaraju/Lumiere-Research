```python
count = 0
num_template = math_templates
for i in num_template:
    num_template[count] = num_template[count].replace("ch", str(q_num[count]))
    num_template[count] = num_template[count].replace("og", str(ans_num[count]))
    num_template[count] = num_template[count].replace(" ", "")
    count += 1
num_template[425020: 425028]
```

```python
finaldf["Number Template"] = num_template
finaldf
```

```python
new_ans = num_template
count = 0
for i in new_ans:
    if ("+" in i):
        new_ans[count] = int(i[0]) + int(i[-1])
        
    if ("-" in i):
        if ("|" in i):
            new_ans[count] = abs(int(i[1]) - int(i[-2]))
        else:
            new_ans[count] = int(i[0]) - int(i[-1])
    
    if ("*" in i):
        new_ans[count] = int(i[0]) * int(i[-1])
    
    if ("/" in i):
        if (int(i[-1]) == 0):
            new_ans[count] == "inf"
        else:
            temp = int(i[0]) / int(i[-1])
            new_ans[count] = round(temp, 1)
    
    if ("//" in i):
        if (int(i[-1]) == 0):
            new_ans[count] == "inf"
        else:
            new_ans[count] = int(i[0]) // int(i[-1])
       
    
    if ("%" in i):
        if (int(i[-1]) == 0):
            new_ans[count] == "inf"
        else:
            new_ans[count] = int(i[0]) % int(i[-1])
    
    
    count += 1
```
